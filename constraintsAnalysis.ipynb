### AREA

```python
import numpy as np
areaArr = np.load("cometsArea.npy")
widthArr = np.load("cometsWidth.npy")
heightArr = np.load("cometsHeight.npy")

areaArrNew = np.reshape(areaArr, -1)
widthArrNew = np.reshape(widthArr, -1)
heightArrNew = np.reshape(heightArr, -1)
print("shape of new areaArr is: ", areaArrNew.shape, widthArrNew.shape, heightArrNew.shape)

areaArrNew, widthArrNew, heightArrNew = areaArr[areaArr != 0], widthArr[widthArr!= 0], heightArr[heightArr!= 0]

#print("\nareaArrNew is: ", areaArrNew)
print("\nshape of new areaArr is: ", areaArrNew.shape, widthArrNew.shape, heightArrNew.shape)

bigArea = areaArr[areaArr>50000]
print("\nbigArea: ",bigArea, "length is: ", len(bigArea))

testY, testX = np.where(areaArr == 2002)
removeIndexBig = np.where(areaArrNew > 50000)  ########## AREA
removeIndexSmall = np.where(areaArrNew < 3000)   ########
removeIndexBig, removeIndexSmall = np.array(removeIndexBig).astype(int), np.array(removeIndexSmall).astype(int)
print("\nremoveIndexes big: ", removeIndexBig) 
print("\nremoveIndexes small: ", removeIndexSmall)
nRemoveIndex = removeIndexBig.shape[1] + removeIndexSmall.shape[1]
print("nRemoveIndex is: ", nRemoveIndex)
removeIndex = np.zeros(nRemoveIndex)
removeIndex[0:removeIndexBig.shape[1]] = removeIndexBig
removeIndex[removeIndexBig.shape[1]:] = removeIndexSmall
print("length of big and small remove indexes are: ", removeIndexBig.shape, removeIndexSmall.shape)
print("length of combined remove index is: ", len(removeIndex))
#print("\nrremoveIndex is: ", removeIndex)

test = np.where(areaArr == 4344)
print("\ntest is: ", test)
print("\n checking an area: ", areaArr[41, 8])

#nRemove = len(removeIndex)

for n in range(nRemoveIndex):
    index = removeIndex[n].astype(int)
    areaArrNew[index] = 0
    widthArrNew[index] = 0
    heightArrNew[index] = 0
    
areaArrNew, widthArrNew, heightArrNew = areaArrNew[areaArrNew != 0], widthArrNew[widthArrNew!= 0], heightArrNew[heightArrNew!= 0]
print("\nArray shapes are now: ", areaArrNew.shape, widthArrNew.shape, heightArrNew.shape)
#print("\nshapes of area, width and height arrays are: ", areaArr.shape, widthArr.shape, heightArr.shape)
print("\nTest Y and test X are: ", testY, testX)

print("removeIndex is: ", removeIndex, "length: ", removeIndex.shape)
print("\n shapes of areaArrNew, widthArrNew and heightArrNew are: ", areaArrNew.shape, widthArrNew.shape, heightArrNew.shape)
print("\nareaArrNew is: ", areaArrNew)
print("\nlength of end area array is: ", areaArrNew.shape)

avgArea = np.average(areaArrNew)
avgWidth = np.average(widthArrNew)
avgHeight = np.average(heightArrNew)
#
minArea = np.min(areaArrNew)
minWidth = np.min(widthArrNew)
minHeight = np.min(heightArrNew)
#
maxArea = np.max(areaArrNew)
maxWidth = np.max(widthArrNew)
maxHeight = np.max(heightArrNew)
#
rangeArea = maxArea - minArea
rangeWidth = maxWidth - minWidth
rangeHeight = maxHeight - minHeight
#
stdevArea = np.std(areaArrNew)
stdevWidth = np.std(widthArrNew)
stdevHeight = np.std(heightArrNew)
#
print("\nArea of comets, min: ", minArea, " max: ", maxArea, " average: ", avgArea, " range: ", rangeArea, " st.dev : ", stdevArea)

print("\nWidth of comets, min: ", minWidth, " max: ", maxWidth, " average: ", avgWidth, " range: ", rangeWidth, " st.dev : ", stdevWidth)

print("\nHeight of comets, min: ", minHeight, " max: ", maxHeight, " average: ", avgHeight, " range: ", rangeHeight, " st.dev : ", stdevHeight)
```
