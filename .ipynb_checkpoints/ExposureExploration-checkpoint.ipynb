### Notebook to test out different methods for getting more of the comet prior to thresholding

```python
import datetime
now = datetime.datetime.now()
print("Date and time ",str(now))
print(" ")
#
import sys
import cv2
import numpy as np
import scipy.ndimage as scimg
import matplotlib.pyplot as plt
%matplotlib inline
!pwd
#
# Read in images
# Numbers ***01.bmp to ...***20.bmp
#imFile = "Images\\ControlBatch\\0001.bmp"
imFile = "../CometImages/CA210609/Collated/4GyB/bmp/4GyB0012.bmp"
#
imgRaw = plt.imread(imFile)
shape = imgRaw.shape
nRows = imgRaw.shape[0] # y coordinates
nCols = imgRaw.shape[1] # x ccordinates
if len(shape) == 3:
    nDepth = imgRaw.shape[2]
else:
    nDepth = 1
imgArr = np.zeros((nRows, nCols))
imgArr[:, :] = imgRaw[:, :]
#
maxPix = np.amax(imgArr)
minPix = np.amin(imgArr)
#
print("\nType of raw image file is",imgRaw.dtype)
print("Type of image file is",imgArr.dtype)
print("\nNumber of rows",nRows,"of columns",nCols,"of pixels",nRows*nCols,"and depth",nDepth)
print("\nMaximum value intensity is: ",maxPix)
print("\nMinimum value intensity is: ",minPix)
#
#thresh = [5,80]
#imgThr = np.zeros((nRows, nCols))
#imgThr = img > thresh[0]
#
imgDouble = imgArr * 2
maxPixDouble = np.amax(imgDouble)
minPixDouble = np.amin(imgDouble)
print("\nmax value of doubled intensities is: ", maxPixDouble)
print("min value of doubled intensities is: ", minPixDouble)
#
fig, ax = plt.subplots(1)
plt.title("Histogram of raw " + str(imFile))
n, bins, patches = ax.hist(imgRaw.flatten(), bins=np.arange(0, maxPix, 1), log=True)
plt.show()
#
fig, ax = plt.subplots(1)
plt.title("Histogram of doubled " + str(imFile))
n, bins, patches = ax.hist(imgDouble.flatten(), bins=np.arange(0, maxPixDouble, 1), log=True)
plt.show()
#
plt.imsave("double4GyB0012.bmp", imgDouble)
plt.imsave("double4GyB0012.png", imgDouble)
cv2.imwrite("cv2Double4GyB0012.bmp", imgDouble)

print("shape of doubled image is: ", imgDouble.shape)

then = now
now = datetime.datetime.now()
print(" ")
print("Date and time",str(now))
print("Time since last check is",str(now - then))
```

## Explore range of gamma values

```python
import datetime
now = datetime.datetime.now()
print("Date and time ",str(now))
print(" ")
#
import sys
import cv2
import numpy as np
import scipy.ndimage as scimg
from skimage import exposure
import matplotlib.pyplot as plt
%matplotlib inline
!pwd
#
# Read in images
# Numbers ***01.bmp to ...***20.bmp
#imFile = "Images\\ControlBatch\\0001.bmp"
imFile = "../CometImages/CA210609/Collated/4GyB/bmp/4GyB0012.bmp"
#
imgRaw = plt.imread(imFile)
shape = imgRaw.shape
nRows = imgRaw.shape[0] # y coordinates
nCols = imgRaw.shape[1] # x ccordinates
if len(shape) == 3:
    nDepth = imgRaw.shape[2]
else:
    nDepth = 1
imgArr = np.zeros((nRows, nCols))
imgArr[:, :] = imgRaw[:, :]
#
maxPix = np.amax(imgArr)
minPix = np.amin(imgArr)
#
print("\nType of raw image file is",imgRaw.dtype)
print("Type of image file is",imgArr.dtype)
print("\nNumber of rows",nRows,"of columns",nCols,"of pixels",nRows*nCols,"and depth",nDepth)
print("\nMaximum value intensity is: ",maxPix)
print("\nMinimum value intensity is: ",minPix)
#
#plot raw file histo
fig, ax = plt.subplots(1)
plt.title("Histogram of raw " + str(imFile))
n, bins, patches = ax.hist(imgArr.flatten(), bins=np.arange(0, maxPix+1, 1), log=True)
plt.show()
#
#subtract min pixel value to get it starting from 0
imgSubtracted = imgArr - minPix
maxPixSub = np.amax(imgSubtracted)
#
#plot subtracted image histo
fig, ax = plt.subplots(1)
plt.title("Histogram of subtracted " + str(imFile))
n, bins, patches = ax.hist(imgSubtracted.flatten(), bins=np.arange(0, maxPixSub+1, 1), log=True)
plt.show()
#
gammaLow = np.linspace(0.1, 1.0, 10 )
gammaHigh = np.linspace(1.1, 2.0, 10)
gammaFine = np.linspace(1.06, 1.15, 10)
print(gammaHigh)
print(gammaFine)
nGamma = len(gammaHigh)
gammaType = 'high fine'
#
if gammaType == 'low':
    for i in range(0, nGamma):
        print("exploring low gamma range (0.1-1.0)")
        print("gain = 1")
        gamma = gammaLow
        imgGamma = exposure.adjust_gamma(imgSubtracted, gamma = gamma[i])
        maxPixGamma = np.amax(imgGamma)
        fig, ax = plt.subplots(1)
        plt.title("Histogram of subtracted and gamma=" + str(gamma[i]) + " transformed " + str(imFile))
        n, bins, patches = ax.hist(imgGamma.flatten(), bins=np.arange(0, maxPixGamma+1, 1), log=True)
        plt.show()
        
elif gammaType == 'high':
    for i in range(0, nGamma):
        print("exploring high gamma range (1.1-2.0)")
        print("gain = 1")
        gamma = gammaHigh
        imgGamma = exposure.adjust_gamma(imgSubtracted, gamma = gamma[i])
        maxPixGamma = np.amax(imgGamma)
        print("max pixel value of adjusted gamma image is: ", maxPixGamma)
        fig, ax = plt.subplots(1)
        plt.title("Histogram of subtracted and gamma=" + str(gamma[i]) + " transformed " + str(imFile))
        n, bins, patches = ax.hist(imgGamma.flatten(), bins=np.arange(0, maxPixGamma+1, 1), log=True)
        plt.show()
elif gammaType == 'high fine':
    for i in range(0, nGamma):
        print("exploring high gamma range (1.1-1.2)")
        print("gain = 1")
        gamma = gammaFine
        imgGamma = exposure.adjust_gamma(imgSubtracted, gamma = gamma[i])
        maxPixGamma = np.amax(imgGamma)
        print("max pixel value of adjusted gamma image is: ", maxPixGamma)
        fig, ax = plt.subplots(1)
        plt.title("Histogram of subtracted and gamma=" + str(gamma[i]) + " transformed " + str(imFile))
        n, bins, patches = ax.hist(imgGamma.flatten(), bins=np.arange(0, maxPixGamma+1, 1), log=True)
        plt.show()
else:
    print("gammaType not understood.")
    sys.exit
    
then = now
now = datetime.datetime.now()
print(" ")
print("Date and time",str(now))
print("Time since last check is",str(now - then))
```

## Explore range of gain values

```python
import datetime
now = datetime.datetime.now()
print("Date and time ",str(now))
print(" ")
#
import sys
import cv2
import numpy as np
import scipy.ndimage as scimg
from skimage import exposure
import matplotlib.pyplot as plt
%matplotlib inline
!pwd
#
# Read in images
# Numbers ***01.bmp to ...***20.bmp
#imFile = "Images\\ControlBatch\\0001.bmp"
imFile = "../CometImages/CA210609/Collated/4GyB/bmp/4GyB0012.bmp"
#
imgRaw = plt.imread(imFile)
shape = imgRaw.shape
nRows = imgRaw.shape[0] # y coordinates
nCols = imgRaw.shape[1] # x ccordinates
if len(shape) == 3:
    nDepth = imgRaw.shape[2]
else:
    nDepth = 1
imgArr = np.zeros((nRows, nCols))
imgArr[:, :] = imgRaw[:, :]
#
maxPix = np.amax(imgArr)
minPix = np.amin(imgArr)
#
print("\nType of raw image file is",imgRaw.dtype)
print("Type of image file is",imgArr.dtype)
print("\nNumber of rows",nRows,"of columns",nCols,"of pixels",nRows*nCols,"and depth",nDepth)
print("\nMaximum value intensity is: ",maxPix)
print("\nMinimum value intensity is: ",minPix)
#
#plot raw file histo
fig, ax = plt.subplots(1)
plt.title("Histogram of raw " + str(imFile))
n, bins, patches = ax.hist(imgArr.flatten(), bins=np.arange(0, maxPix+1, 1), log=True)
plt.show()
#
#subtract min pixel value to get it starting from 0
imgSubtracted = imgArr - minPix
maxPixSub = np.amax(imgSubtracted)
#
#plot subtracted image histo
fig, ax = plt.subplots(1)
plt.title("Histogram of subtracted " + str(imFile))
n, bins, patches = ax.hist(imgSubtracted.flatten(), bins=np.arange(0, maxPixSub+1, 1), log=True)
plt.show()
#
gain = np.linspace(1.0, 2.0, 11)
print(gain)
nGain = len(gain)
#
for i in range(0, nGain):
    print("exploring gain values")
    print("gamma = 1")
    imgGain = exposure.adjust_gamma(imgSubtracted, gain = gain[i])
    maxPixGain = np.amax(imgGain)
    print("max pixel value after adjusting gain is: ", maxPixGain)
    fig, ax = plt.subplots(1)
    plt.title("Histogram of subtracted and gain=" + str(gain[i]) + " transformed " + str(imFile))
    n, bins, patches = ax.hist(imgGain.flatten(), bins=np.arange(0, maxPixGain+1, 1), log=True)
    plt.show()

then = now
now = datetime.datetime.now()
print(" ")
print("Date and time",str(now))
print("Time since last check is",str(now - then))
```

## compare the original intensity histogram with the gamma=1.1 and gain=1.6 histograms

```python
import datetime
now = datetime.datetime.now()
print("Date and time ",str(now))
print(" ")
#
import sys
import cv2
import numpy as np
import scipy.ndimage as scimg
from skimage import exposure
import matplotlib.pyplot as plt
%matplotlib inline
!pwd
#
# Read in images
# Numbers ***01.bmp to ...***20.bmp
#imFile = "Images\\ControlBatch\\0001.bmp"
imFile = "../CometImages/CA210609/Collated/4GyB/bmp/4GyB0012.bmp"
#
imgRaw = plt.imread(imFile)
shape = imgRaw.shape
nRows = imgRaw.shape[0] # y coordinates
nCols = imgRaw.shape[1] # x ccordinates
if len(shape) == 3:
    nDepth = imgRaw.shape[2]
else:
    nDepth = 1
imgArr = np.zeros((nRows, nCols))
imgArr[:, :] = imgRaw[:, :]
#
maxPix = np.amax(imgArr)
minPix = np.amin(imgArr)
print("min and max pixel values of raw image are: ", minPix, maxPix)
#
imgSubtracted = imgArr - minPix
imgGain = exposure.adjust_gamma(imgSubtracted, gain = 1.6)
imgGamma = exposure.adjust_gamma(imgSubtracted, gamma = 1.1)
maxPixGain = np.amax(imgGain)
maxPixGamma = np.amax(imgGamma)
print("max pixel value after adjusting gain is: ", maxPixGain)
print("max pixel value after adjusting gamma is: ", maxPixGamma)
#
#plot histograms of raw, gamma and gain adjusted image
fig, ax = plt.subplots(1, figsize = (16, 12))
plt.title("Histogram of raw(subtracted, gamma and gain adjusted image " + str(imFile))
n, bins, patches = ax.hist(imgSubtracted.flatten(), bins=np.arange(0, maxPix+1, 1), alpha = 0.5, log=True, label='raw subtracted')
n, bins, patches = ax.hist(imgGain.flatten(), bins=np.arange(0, maxPixGain+1, 1), alpha = 0.5, log=True, label='gain adjusted')
n, bins, patches = ax.hist(imgGamma.flatten(), bins=np.arange(0, maxPixGamma+1, 1), alpha = 0.5, log=True, label='gamma adjusted')
ax.legend(loc='best', frameon=False)
plt.show()
#
#save gain and gamma adjusted images to file
cv2.imwrite("gamma4GyB0012.bmp", imgGamma)
cv2.imwrite("gain4GyB0012.bmp", imgGain)
#
then = now
now = datetime.datetime.now()
print(" ")
print("Date and time",str(now))
print("Time since last check is",str(now - then))
```
